```python
backup = df.copy()
```

```python
df['runs'] = pd.to_numeric(df['runs'], errors = 'coerce')
```

```python
df['runs'].dtype
```

```python
df['current_score'] = df.groupby('match_id')['runs'].cumsum()
```

```python
df
```

```python
backup2 = df.copy()
```

```python
# balls left
```

```python
df['over'] = df['ball'].apply(lambda x:str(x).split('.')[0])
df['ball_no'] = df['ball'].apply(lambda x:str(x).split('.')[1])
```

```python
df
```

```python
df['balls_bowled'] = (df['over'].astype('int')*6) + df['ball_no'].astype('int')
```

```python
df
```

```python
df['balls_left'] = 120 - df['balls_bowled']
df['balls_left'] = df['balls_left'].apply(lambda x:0 if x<0 else x)
```

```python
df
```

```python
backup3 = df.copy()
```

```python
 # wickets left
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
```

```python
df
```

```python
df['player_dismissed'] = df['player_dismissed'].astype('int')
df['player_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()
df['wickets_left'] = 10 - df['player_dismissed']
```

```python
df
```

```python
backup3 = df.copy()
```

```python
df['crr'] = round((df['current_score']*6)/df['balls_bowled'],2)
```

```python
df
```

```python
groups = df.groupby('match_id')

match_ids = df['match_id'].unique()
last_five = []
for id in match_ids:
    # Get the group corresponding to the current match ID
    group = groups.get_group(id)
    
    # Ensure 'runs' column is numeric
    group['runs'] = pd.to_numeric(group['runs'], errors='coerce')
    
    # Apply rolling sum with window of 30 and get the 'runs' column values
    rolling_sums = group['runs'].rolling(window=30).sum().values
    
    # Extend the last_five list with the rolling sums
    last_five.extend(rolling_sums.tolist())
```

```python
df['last_five'] = last_five
```

```python
df
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
final_df = final_df[['batting_team','bowling_team','current_score','wickets_left','crr','city','balls_left','last_five','runs_x']]
```

```python
X = final_df.drop(columns = ['runs_x'])
y = final_df['runs_x']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```
